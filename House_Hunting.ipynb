Выбор квартиры (House-hunting)

Обоснование: набор данных представляет собой критерии для выбора квартиры, где "ID" - номер жилья, "DISTANCE" - расстояние от метро пешком(в минутах), "STOP_COUNT"-количество остановок до места учебы/работы, "COST" - ежемесячная цена аренды(в тысячах рублей). Необходимо выбрать три наилучших варианта с наименьшей стоимостью, наименьшим количеством времени до метро и минимальным количеством проезжаемых станций.


```python
import pandas as pd
data = pd.read_csv('https://courses.openedu.ru/assets/courseware/v1/56e6a86e4d025ce9847e52b375f8ac6b/asset-v1:ITMOUniversity+DATANTECH2035+fall_2021_3+type@asset+block/task4_372666.csv',
                   delimiter=',',
                   decimal='.',
                   index_col='ID')
data
```

Выполнение экспоненциальной нормировки для выбранных данных.

```python
import numpy as np
data = 1 - np.exp(1 - data/data.min())
data
```

Создание дополнительного столбца со значениями целевой функции

```python
data['SUM'] = 0
data['SUM'] = data['DISTANCE'] + data['STOP_COUNT'] + data['COST']
```

Выведем три лучших варианта

```python
def find_best_variant():
    result = data.sort_values(by = 'SUM')[0:3]
    return result
find_best_variant()
```

В результате было получено, что наилучшие варианты квартир имеют ID 6,123 и 39.
